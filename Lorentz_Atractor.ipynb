Atractor de Lorentz

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0

def f(state, t):
    x, y, z = state  # Desempaqueta el vector de estado
    return sigma * (y - x), x * (rho - z) - y, x * y - beta * z  # Derivadas

state0 = [1.0, 1.0, 1.0]
t = np.arange(0.0, 40.0, 0.01)

states = odeint(f, state0, t)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot(states[:, 0], states[:, 1], states[:, 2])

fig.set_size_inches(12, 8)
plt.show()

fig.savefig('lorenz_attractor.png')
```

```python
# Parámetros del Atractor de Rössler
a = 0.2
b = 0.2
c = 5.7

# Definición del sistema de ecuaciones diferenciales
def rossler(state, t):
    x, y, z = state
    dxdt = -y - z
    dydt = x + a * y
    dzdt = b + z * (x - c)
    return [dxdt, dydt, dzdt]

# Condiciones iniciales
state0_rossler = [1.0, 1.0, 1.0]

# Tiempo de integración
t_rossler = np.arange(0.0, 100.0, 0.01)

# Resolución del sistema de ecuaciones diferenciales
states_rossler = odeint(rossler, state0_rossler, t_rossler)

# Gráfica del Atractor de Rössler
fig_rossler = plt.figure()
ax_rossler = fig_rossler.add_subplot(projection='3d')
ax_rossler.plot(states_rossler[:, 0], states_rossler[:, 1], states_rossler[:, 2])

fig_rossler.set_size_inches(12, 8)
plt.show()

fig.savefig('rossler_attractor.png')
```
